```python
def replace_x(df):

    for col in df.columns:
        
        new_col = col.replace("_x", " 2023")
        
        df.rename(columns={col: new_col}, inplace=True)
        
    print(df)

```

```python
def replace_y(df):

    for col in df.columns:
        
        new_col = col.replace("_y", " 2022")
        
        df.rename(columns={col: new_col}, inplace=True)
        
    print(df)
```

```python
replace_x(wpi)
```

```python
replace_y(wpi)
```
